```python
import random

questions = [
    {"question": "How many hearts does an octopus have?",
     "options": ["A) 1", "B) 3", "C) 9"], "answer": "B"},

    {"question": "Where does the Python language get its name from?",
     "options": ["A) From the snake", "B) From the Monty Python comedy group", "C) From a movie"], "answer": "B"},

    {"question": "Why does one sock always disappear in the washing machine?",
     "options": ["A) It goes to a parallel universe", "B) It evaporates", "C) We'll never know"], "answer": "A"},

    {"question": "Does a giraffe have more neck vertebrae than a human?",
     "options": ["A) Yes, many more", "B) No, both have 7", "C) Giraffes have none"], "answer": "B"},

    {"question": "How long does a phone last at 1% battery?",
     "options": ["A) 1 second", "B) Forever", "C) Exactly until you need it most"], "answer": "C"},
]

game_state = {"score": 0, "current_question": 0, "game_over": False}


def display_welcome():
    print("=" * 35)
    print("🎯 WELCOME TO THE QUIZ GAME!")
    print(f"There are {len(questions)} questions. Good luck!")
    print("=" * 35)


def get_valid_choice():
    while True:
        try:
            choice = input("Your answer (A/B/C): ").strip().upper()
            if choice in ["A", "B", "C"]:
                return choice
            else:
                print("⚠️ Invalid! Please enter only A, B or C.")
        except Exception:
            print("⚠️ Something went wrong, please try again.")


def ask_question(q):
    print("\n" + q["question"])
    for option in q["options"]:         
        print(option)

    choice = get_valid_choice()

    if choice == q["answer"]:           
        print("✅ Correct!")
        return True
    else:
        print(f"❌ Wrong! The correct answer is: {q['answer']}")
        return False


def play_quiz():
    display_welcome()
    random.shuffle(questions)            

    for q in questions:                 
        if ask_question(q):
            game_state["score"] += 1    
        game_state["current_question"] += 1

    game_state["game_over"] = True
    show_result()


def show_result():
    score = game_state["score"]
    total = len(questions)
    print("\n" + "=" * 35)
    print(f"Game over! Your score: {score}/{total}")

    if score >= total / 2:              
        print("🎉 YOU WIN! Congratulations!")
    else:                              
        print("💀 YOU LOSE! Try again.")
    print("=" * 35)


play_quiz()
```
